```python
%matplotlib inline
import matplotlib.pyplot as plt

route = []
with open("outputs/route_SA13.dat") as fp:
    for line in fp.readlines():
        string = line.split(' ')
        route.append(string)
for i in range(3):
    route.pop(0)
x_route = []
y_route = []
for i in route:
    x_route.append(i[1])
    y_route.append(i[2])
x_route.append(route[0][1])
y_route.append(route[0][2])
plt.plot(x_route, y_route)
```
